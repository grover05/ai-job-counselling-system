# 🎯 Career Guidance Chatbot - Structured Response Format

This notebook demonstrates the new **structured response format** from the improved career guidance chatbot API.

Instead of returning a single text block, the chatbot now returns organized JSON with clear sections:
- Career Overview
- Required Skills
- Recommended Tools
- Suggested Projects
- Step-by-Step Learning Path
- Actionable Suggestions

This makes it much easier for the frontend to display each section beautifully!

## Setup & Configuration

````python
import requests
import json
from pprint import pprint

# API Configuration
BACKEND_URL = "http://127.0.0.1:8000"
STRUCTURED_ENDPOINT = f"{BACKEND_URL}/api/chatbot/query-structured"
REGULAR_ENDPOINT = f"{BACKEND_URL}/api/chatbot/query"

print(f"✅ Backend: {BACKEND_URL}")
print(f"✅ Structured Endpoint: /api/chatbot/query-structured")
print(f"✅ Regular Endpoint: /api/chatbot/query")
````

## Example 1: AI/ML Engineer Career Path (Structured Format)

````python
# Query for AI/ML Engineer
query = {"message": "I want to become an AI/ML engineer"}

response = requests.post(STRUCTURED_ENDPOINT, json=query)
structured_data = response.json()

print("📊 STRUCTURED RESPONSE FORMAT:")
print("=" * 80)
pprint(structured_data, depth=5)
````

## Breaking Down the Structured Response

````python
# Access individual sections
print("\n🎯 CAREER ROLE DETECTED:")
print(f"   Role: {structured_data['role_detected']}")
print(f"   Status: {structured_data['status']}")

print("\n📋 CAREER OVERVIEW:")
overview = structured_data.get('overview', {})
print(f"   Title: {overview.get('title')}")
print(f"   Description: {overview.get('description')}")

print("\n💼 REQUIRED SKILLS:")
skills = structured_data.get('skills', {}).get('items', [])
for i, skill in enumerate(skills, 1):
    print(f"   {i}. {skill}")

print("\n🛠️ RECOMMENDED TOOLS:")
tools = structured_data.get('tools', {}).get('items', [])
for i, tool in enumerate(tools, 1):
    print(f"   {i}. {tool}")
````

````python
print("\n🚀 SUGGESTED PROJECTS:")
projects = structured_data.get('projects', {}).get('items', [])
for i, project in enumerate(projects, 1):
    print(f"   {i}. {project}")

print("\n📚 LEARNING PATH (Step-by-Step):")
learning_path = structured_data.get('learning_path', {}).get('steps', [])
for step in learning_path:
    print(f"\n   Step {step['step_number']}: {step['title']}")
    if step['description']:
        print(f"   Description: {step['description'][:80]}...")

print("\n💡 SUGGESTIONS:")
suggestions = structured_data.get('suggestions', [])
for suggestion in suggestions:
    print(f"   • {suggestion}")
````

## Example 2: Data Scientist Career Path

````python
# Query for Data Scientist
query = {"message": "How do I become a data scientist?"}

response = requests.post(STRUCTURED_ENDPOINT, json=query)
data_scientist_response = response.json()

print("\n🎯 CAREER ROLE: Data Scientist")
print("=" * 80)

overview = data_scientist_response.get('overview', {})
print(f"\n📋 {overview.get('title')}")
print(f"{overview.get('description')}")

print(f"\n💼 Required Skills ({len(data_scientist_response.get('skills', {}).get('items', []))} total):")
for i, skill in enumerate(data_scientist_response.get('skills', {}).get('items', [])[:5]):
    print(f"   {i+1}. {skill}")
print(f"   ... and {len(data_scientist_response.get('skills', {}).get('items', [])) - 5} more")
````

## Example 3: Greetings & Special Cases

````python
# Test greeting
query = {"message": "Hello! Can you help me?"}
response = requests.post(STRUCTURED_ENDPOINT, json=query)
greeting_response = response.json()

print("\n👋 GREETING RESPONSE:")
print(f"Role Detected: {greeting_response['role_detected']}")
print(f"Message: {greeting_response.get('message')}")

# Test unclear query
print("\n" + "="*80)
query = {"message": "What should I do?"}
response = requests.post(STRUCTURED_ENDPOINT, json=query)
unclear_response = response.json()

print("\n❓ UNCLEAR QUERY RESPONSE:")
print(f"Is Clarification: {unclear_response['is_clarification']}")
print(f"Role Detected: {unclear_response['role_detected']}")
print(f"\nMessage:\n{unclear_response.get('message')}")
````

## Comparison: Regular vs Structured Format

````python
query = {"message": "Tell me about backend development"}

# Regular format
print("=" * 80)
print("📝 REGULAR FORMAT (Plain Text):")
print("=" * 80)
regular_response = requests.post(REGULAR_ENDPOINT, json=query).json()
print(regular_response['message'][:500] + "...\n")

# Structured format
print("=" * 80)
print("📊 STRUCTURED FORMAT (JSON):")
print("=" * 80)
structured_response = requests.post(STRUCTURED_ENDPOINT, json=query).json()
print(f"Status: {structured_response['status']}")
print(f"Role: {structured_response['role_detected']}")
print(f"Overview: {structured_response['overview']['title']}")
print(f"Skills: {len(structured_response['skills']['items'])} items")
print(f"Tools: {len(structured_response['tools']['items'])} items")
print(f"Projects: {len(structured_response['projects']['items'])} items")
print(f"Learning Steps: {len(structured_response['learning_path']['steps'])} steps")
print(f"Suggestions: {len(structured_response['suggestions'])} actionable items")
````

## 🎨 Frontend Usage Guide

### Use the structured format for better UI display:

```javascript
// Frontend code example
async function getCareerGuidance(message) {
  const response = await fetch('http://localhost:8000/api/chatbot/query-structured', {
    method: 'POST',
    headers: { 'Content-Type': 'application/json' },
    body: JSON.stringify({ message })
  });
  
  const data = await response.json();
  
  // Now you can easily display each section:
  displayCareerOverview(data.overview);      // Title + Description
  displaySkillsList(data.skills.items);      // Numbered skill list
  displayToolsList(data.tools.items);        // Numbered tools list
  displayProjects(data.projects.items);      // Project ideas
  displayLearningPath(data.learning_path.steps);  // Step-by-step guide
  displaySuggestions(data.suggestions);      // Action items
}
```

### Benefits of Structured Format:
✅ **Easier Frontend Integration** - Each section is clearly separated
✅ **Better Mobile Display** - Easy to create collapsible sections
✅ **Improved Accessibility** - Semantic structure for screen readers
✅ **Better Styling** - Can apply different CSS to each section type
✅ **Easier Data Processing** - No need to parse text
✅ **Flexible Rendering** - Can display sections in any order

## 🎉 Summary

The **structured response format** provides:

1. **Clear Organization** - Each section has its own field
2. **Better Type Safety** - Predictable JSON structure
3. **Easier Frontend Integration** - No parsing needed
4. **Improved UX** - Can display sections with proper styling
5. **Mobile Friendly** - Easy to create responsive layouts

### API Endpoints:
- **Regular Format**: `POST /api/chatbot/query`
  - Returns: Plain text response
  - Use for: Quick text display

- **Structured Format**: `POST /api/chatbot/query-structured`
  - Returns: Organized JSON with sections
  - Use for: Advanced UI with multiple sections

Both endpoints return the same brilliant career guidance, just in different formats! 🚀
